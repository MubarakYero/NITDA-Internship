```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.scrapethissite.com/pages/simple/'
```

```python
page_to_scrape = requests.get(url)
```

```python
soup = BeautifulSoup(page_to_scrape.text, 'html')
```

```python
print(soup)
```

### Extracting Columns

```python
columns = soup.find_all('strong')[:3]

columns_data = [data.text.strip() for data in columns]

print(columns_data)
```

```python
cleaned_columns = [text.rstrip(':') for text in columns_data]
cleaned_columns
```

#### Adding Country Column               

```python
cleaned_columns.insert(0, 'Country')
cleaned_columns
```

### Extracting Country Data

```python
countries = soup.find_all('h3', class_="country-name")

print(countries)
```

```python
countries_name = [name.text.strip() for name in countries]

print(countries_name)
```

### Extracting Capital Data

```python
capitals = soup.find_all('span', class_="country-capital")

print(capitals)
```

```python
capital_names = [name.text.strip() for name in capitals]

print(capital_names)
```

### Extracting Population Data

```python
population = soup.find_all('span', class_="country-population")

print(population)
```

```python
population_data = [data.text.strip() for data in population]

print(population_data)
```

```python
population_data = [text.rstrip('[]') for text in population_data]
population_data
```

### Extracting Size Data

```python
area = soup.find_all('span', class_="country-area")

print(area)
```

```python
area_size = [size.text.strip() for size in area]

print(area_size)
```

### Adding data into a DataFrame

```python
import pandas as pd
```

```python
countries_data = pd.DataFrame(columns=cleaned_columns)

countries_data
```

### Adding Rows to dataframe

```python
for i in range(len(countries_name)):
    countries_data.loc[i] = [countries_name[i], capital_names[i], population_data[i], area_size[i]]
```
